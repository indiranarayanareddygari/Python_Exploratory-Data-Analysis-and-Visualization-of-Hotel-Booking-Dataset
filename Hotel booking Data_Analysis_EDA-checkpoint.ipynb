Data analysis and Visualisations

```python
cancelled_prec = df['is_canceled'].value_counts(normalize = True)
print(cancelled_prec)
```

```python
#visualisations
# we will plot bar graph because data which we are presenting gives u categorical and
# we are going to present some count
plt.figure(figsize = (5,4))
plt.title('Reservation_Status_Count')
plt.bar(['Not_canceled','canceled'],df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
# Replacing 0 with 'not canceled' and 1 with 'canceled' in the 'is_canceled' column
df['is_canceled'] = df['is_canceled'].replace({0: 'not canceled', 1: 'canceled'})

# Plotting the count of reservation status in different hotels
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
ax1.legend(bbox_to_anchor=(1, 1))
legend_labels, _ = ax1.get_legend_handles_labels()
plt.title('Reservation Status in different hotels', size=20)
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
#the code segment identifies the cancellation status distribution (canceled vs. not canceled) 
# within the bookings specifically categorized as 'Resort Hotel', presenting the percentage or
# proportion of each cancellation status
```

```python
City_hotel = df[df['hotel'] == 'City Hotel']
City_hotel['is_canceled'].value_counts(normalize = True)
# this code segment identifies the distribution of cancellation statuses (canceled vs. not canceled) 
# within the bookings categorized specifically as 'City Hotel', presenting the percentage or 
# proportion of each cancellation status.
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()

City_hotel = City_hotel.groupby('reservation_status_date')[['adr']].mean()

#These code snippets calculate the average daily rate ('adr') 
# grouped by reservation dates separately for 'Resort Hotel' and 'City Hotel' bookings.
```

```python
# # Creating a visualization to compare the trend of Average Daily Rates between 
# Resort and City Hotels using matplotlib.pyplot 
# in a figure of size 20x8 inches with titles and labels

plt.figure(figsize=(20,8))

plt.title('Average Daily Rate in City and Resort Hotel',fontsize = 30)
plt.plot(resort_hotel.index,resort_hotel['adr'], label = 'resort hotel')

plt.plot(City_hotel.index,City_hotel['adr'], label = 'resort hotel')
```

```python
# Visualization depicting the count of reservations categorized by cancellation status per month, 
# using Seaborn's countplot with legends, titles, and labels included.

df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x= 'month',hue = 'is_canceled',data = df, palette= 'bright')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation Status Per Month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not cancelled','canceled'])
plt.show()

```

```python
import matplotlib.pyplot as plt

# Convert 'not canceled' to 0 and 'canceled' to 1 in 'is_canceled' column
df['is_canceled'] = df['is_canceled'].replace({'not canceled': 0, 'canceled': 1})

# Grouping by month and counting reservations
total_reservations = df.groupby('month').size().reset_index(name='total_reservations')

# Grouping by month and counting canceled reservations
canceled_reservations = df[df['is_canceled'] == 1].groupby('month').size().reset_index(name='canceled_reservations')

# Merging the two DataFrames on 'month'
reservation_stats = total_reservations.merge(canceled_reservations, on='month')

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(reservation_stats['month'], reservation_stats['total_reservations'], label='Total Reservations')
plt.bar(reservation_stats['month'], reservation_stats['canceled_reservations'], label='Canceled Reservations')

plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.title('Total vs Canceled Reservations per Month')
plt.legend()
plt.xticks(reservation_stats['month'])  # Ensure all months are shown on x-axis
plt.show()


##This code visualizes the total reservations against canceled reservations per month using a bar plot. 
# It computes the counts for both total reservations and canceled reservations, grouped by month, 
# and plots the data for comparison
```

```python
# Visualizes the average daily rate per month using Seaborn's barplot based on the calculated mean daily rates grouped by month.

import seaborn as sns
import matplotlib.pyplot as plt

# Calculate average daily rate for each month
average_daily_rate = df.groupby('month')['adr'].mean().reset_index()

# Plotting average daily rate for each month with Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='adr', data=average_daily_rate, palette='viridis')
plt.xlabel('Month')
plt.ylabel('Average Daily Rate')
plt.title('Average Daily Rate per Month')
plt.show()
```

```python
#Generates a pie chart displaying the top 10 countries with the highest number of canceled reservations from the dataset.

Cancelled_data = df[df['is_canceled'] == 1]
top_10_Country = Cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 Countries with reservation cancelled')
plt.pie(top_10_Country,autopct = '%.2f', labels = top_10_Country.index)
plt.show()
```

```python
#The code calculates and compares the average daily rate over time between canceled and not-canceled reservations, 
# showcasing the trend of the daily rates for both categories on a time-series plot.

cancelled_data = df[df['is_canceled'] == 1]
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
plt.legend()
plt.show()
```
